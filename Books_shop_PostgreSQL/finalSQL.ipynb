## Решение заданий:
### Рассмотрим, сколько книг вышло после 1 января 2000 года;

```python
get_sql_data( '''
SELECT COUNT(book_id)
FROM books 
WHERE publication_date > '2000-01-01'
''')
```

Всего 819 книг вышло после 01.01.2000 года

### Для каждой книги посчитаем количество обзоров и среднюю оценку;

```python
get_sql_data( '''
WITH 
    rat AS (
            SELECT book_id,
            ROUND(AVG(rating), 2) AS avg_rating
            FROM ratings 
            GROUP BY book_id
            ),

    rev AS (
            SELECT book_id,
            COUNT(review_id) AS cnt_reviews
            FROM reviews 
            GROUP BY book_id
            )
SELECT b.book_id,
       b.title,
       rev.cnt_reviews,
       rat.avg_rating            
FROM books AS b
JOIN rev ON rev.book_id = b.book_id
JOIN rat ON rat.book_id = b.book_id
ORDER BY rat.avg_rating DESC
       ''')
```

```python
get_sql_data( '''
           SELECT b.book_id, 
                  b.title,
                  COUNT(DISTINCT rw.review_id) AS n_reviews, 
                  ROUND(AVG(rt.rating), 2) AS avg_rating
           FROM books AS b
           LEFT OUTER JOIN reviews AS rw ON b.book_id = rw.book_id
           LEFT OUTER JOIN ratings AS rt ON b.book_id = rt.book_id
           GROUP BY b.book_id
           ORDER BY avg_rating DESC, n_reviews DESC
        ''')
```

Всего 1000 книг имеют обзоры и рейтинги.<br>

- Welcome to Temptation.

### Определим издательство, которое выпустило наибольшее число книг толще 50 страниц

```python
get_sql_data('''
      SELECT p.publisher,
                   COUNT(b.book_id) AS total_books
            FROM books AS b
            JOIN publishers AS p ON b.publisher_id = p.publisher_id
            WHERE b.num_pages > 50
            GROUP BY p.publisher_id
            ORDER BY total_books DESC
            LIMIT 1
              ''')
```

Издательство Penguin Books выпускает книги толще 50 страниц с числом книг в 42шт.

### Определим автора с самой высокой средней оценкой книг — (только книги с 50 и более оценками)

```python
get_sql_data('''
SELECT author,
       ROUND(AVG(rating), 2) as avg
FROM authors AS a
INNER JOIN books AS b ON b.author_id=a.author_id
INNER JOIN ratings AS r ON r.book_id=b.book_id
WHERE b.book_id IN (SELECT  book_id
                            FROM ratings
                            GROUP BY book_id
                            HAVING COUNT(rating) >= 50)
GROUP BY author
ORDER BY AVG(rating) DESC
LIMIT 1
             ''')
```

Самый высокорейтинговый автор J.K. Rowling/Mary GrandPré со средней оценкой 4,29

### Среднее количество обзоров от пользователей, которые поставили больше 48 оценок.

```python
get_sql_data('''
        WITH 
            temp AS (
                     SELECT username,
                     COUNT(review_id) AS cnt
                     FROM reviews
                     WHERE username IN (
                                        SELECT username
                                        FROM ratings
                                        GROUP BY username
                                        HAVING COUNT(rating) > 48
                                        )
                     GROUP BY username
                     )
SELECT ROUND(AVG(cnt), 2) AS count_rewiew
FROM temp
             ''')
```

Среднее количество обзоров пользователей, который поставили больше 48 оценок - 24.
